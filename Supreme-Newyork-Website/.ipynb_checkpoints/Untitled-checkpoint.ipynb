```python
import requests
import json
import pandas as pd
import sqlite3


response = requests.get("https://www.supremenewyork.com/mobile_stock.json")
response_list = json.loads(response.text)

```

```python
response_keys = []

for items in response_list:
    response_keys.append(items)

print(response_keys)
```

```python
Date = response_list[response_keys[-2]]
Fashion_Week = response_list[response_keys[-1]]
print(Fashion_Week)
print(Date)
```

```python
items_info_list = response_list[response_keys[1]]

TABLE = []

for category in items_info_list.keys():
    print("----")
    print(category)
    print("----")
    for values in items_info_list[category]:
        RECORD = []
        for a,b in values.items():
            RECORD.append(b)
        print(RECORD)
        TABLE.append((category,RECORD[0],RECORD[1],RECORD[2],RECORD[3],RECORD[4],RECORD[5],RECORD[6],RECORD[7],RECORD[8]))


```

```python
TABLE = pd.DataFrame(TABLE , columns = ['Main_Category','Name', 'Product_Id','Image_Url','Image_Url_Hi','Price','Sale_Price','New_Item',
                                       'Position','Category_Name']) 
TABLE["Date"] = Date
TABLE["Fashion_Week"] = Fashion_Week

TABLE = TABLE[[
    'Date',
    'Fashion_Week',
    'Main_Category',
    'Category_Name',
    'Name',
    'Product_Id',
    'Price',
    'Sale_Price',
    'New_Item',
    'Position'
]]
```
